# 크롤링 (lxml + selenium)

네이버 뮤직 윤딴딴 모든 노래 가사 크롤링

```python
import requests  ###서버에 요청을 할떄 쓰는 모듈.
import lxml.html #xml 방식으로 파싱할떄 쓰이는 모듈 
import datetime as dt
url= "https://music.naver.com/artist/track.nhn?artistId=319146&page=1"
# 유저 에이전트
res = requests.get(url, headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) '
                   'Chrome/41.0.2228.0 Safari/537.36')})
res.status_code
```

```python
# HTML 소스 가져오기
html = res.text   ####아까 content와 같은 응답내용이지만 text형식으로 표현됨.
html
```

```python
e = lxml.html.fromstring(res.content) ####파싱하는  모듈 (분석)
e
```

```python
href = e.cssselect('.btn_lyrics') #css 선택자로 .btn_lyrics이라는 태그를 가진
```

```python
import re
```

```python
url_l = list()
p1 = re.compile('\d+')
for i in href:
    href_num = p1.findall(i.get('onclick'))[0]
    
    url_s = 'https://music.naver.com/lyric/index.nhn?trackId=' + href_num
    res_s = requests.get(url_s, headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) '
                   'Chrome/41.0.2228.0 Safari/537.36')})
    e_s = lxml.html.fromstring(res_s.content)
    lyric = e_s.cssselect('#lyricText br')
    if lyric:
        url_l.append(url_s)   
```

```python
len(url_l)
```

```python
from selenium.webdriver import Chrome
from time import sleep
browser=Chrome()
```

```python
temp = list()
for i in url_l:
    browser.get(i)
    comments=browser.find_elements_by_css_selector('#lyricText')
    temp.append(comments[0].text.split('\n'))
    sleep(1)
```

```python
lyric_l = list()
for i in temp:
    lyric_l.append([x for x in i if x])
```

```python
lyric_l
```

# skip gram model

```python
import numpy as np
import pandas as pd
from konlpy.tag import *
twitter = Twitter()
```

```python
def make_corpus_rm_stopwords(text):
    corpus = []
    for line in text:
        for s in line:
            corpus.append(['/'.join(p) for p in twitter.pos(s) if p[1] != "Josa"])
    return corpus
```

```python
corpus = make_corpus_rm_stopwords(lyric_l)
```

```python
corpus
```

```python
from gensim.models import Word2Vec, fasttext
```

```python
Skip_Gram_model = Word2Vec(corpus, size=2, window=2, min_count=2, workers=1, iter=1000, sg=1)
```

```python
words = Skip_Gram_model.wv.index2word
words
```

```python
vectors = Skip_Gram_model.wv.vectors
vectors
```

```python
Skip_Gram_model_result = dict(zip(words, vectors))
```

```python
Skip_Gram_model.most_similar('마음/Noun', topn=5)
```
